```python
import pandas as pd

def sample_equal_rows_per_class(df, target_column, n=20, random_state=42):
    """
    Samples `n` rows for each class in the target column.
    
    Parameters:
    - df: DataFrame
    - target_column: str, name of the target column
    - n: int, number of samples per class
    - random_state: int, for reproducibility

    Returns:
    - DataFrame with n rows per class
    """
    return df.groupby(target_column, group_keys=False).apply(
        lambda x: x.sample(n=min(len(x), n), random_state=random_state)
    ).reset_index(drop=True)

```

```python
path=r"UPI Transactions 2024 Dataset.csv"
df = pd.read_csv(path)
balanced_df = sample_equal_rows_per_class(df, target_column='fraud_flag')
filename=r"Test_data.csv"
balanced_df.to_csv(filename, index=False)
```
